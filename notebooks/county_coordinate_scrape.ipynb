```python
import urllib.request
import pandas as pd
import requests
from bs4 import BeautifulSoup
import gzip
import os
from time import sleep
```

```python
response = requests.get(url="https://en.wikipedia.org/wiki/User:Michael_J/County_table")
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table')
rows = table.findAll('tr')
```

```python
fips = []
state = []
county = []
latitude = []
longitude = []

def degree_processing(txt):
    if txt[0] == 'â€“':
        return round(-float(txt[1:-2]),3)
    else:
        return round(float(txt[:-2]),3)

def text_processing(txt):
    if txt[-1] == ']':
        txt = txt[:-4]
    if txt[:7] == 'Georgia':
        return 'Georgia'
    return txt.rstrip(' ')

for row in rows[1:]:
    tds = row.findAll('td')
    fips.append(str(tds[2].text))
    state.append(text_processing(tds[1].find('a')['title']))
    county.append(text_processing(tds[3].text))
    latitude.append(degree_processing(tds[-2].text))
    longitude.append(degree_processing(tds[-1].text))
```

```python
ad_hoc_ct = [ ['09110','Capitol Planning Region', 41.51, -72.39], ['09120', 'Greater Bridgeport Planning Region', 41.11, -73.11], ['09130', 'Lower Connecticut River Valley Planning Region', 41.28, -72.31], 
['09140', 'Naugatuck Valley Planning Region', 41.31, -72.03], ['09150', 'Northeastern Connecticut Planning Region', 41.47, -71.56], ['09160', 'Northwest Hills Planning Region', 41.50, -73.13],
['09170', 'South Central Connecticut Planning Region', 41.22, 72.49], ['09180','Southeastern Connecticut Planning Region', 41.29, 72.05], ['09190', 'Western Connecticut Planning Region', 41.19, 73.29]   ]



for f, ct, lt, lo in ad_hoc_ct:
    fips.append(f)
    state.append('Connecticut')
    county.append(ct)
    latitude.append(lt)
    longitude.append(lo)
```

```python
ad_hoc_last = [['12025','Florida','Dade',  25.55, -80.63], ['46113', 'South Dakota', 'Shannon', 43.33, -102.55], ['51780', 'Virginia', 'South Boston', 36.42, -78.54]]


for f, s, ct, lt, lo in ad_hoc_last:
    fips.append(f)
    state.append(s)
    county.append(ct)
    latitude.append(lt)
    longitude.append(lo)
```

```python
df = pd.DataFrame({"fips":fips, "state": state, "county": county, "latitude": latitude, "longitude": longitude})
df.to_csv("county_coordinate.csv")
```
